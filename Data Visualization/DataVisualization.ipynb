## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)  #Random Data 
```

```python
plt.plot(x,y1,color="red",label="Apple",marker="o")
plt.plot(x,y2,label="Kiwi",linestyle="dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

### Scatter Plots

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color="red",label="Apple")
plt.scatter(x,y2,label="Kiwi",color="green",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.figure(figsize=(10,10))
plt.show()
```

### Bar Graphs

```python
x_corr = np.array([0,1,2])*2
plt.bar(x_corr,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"])
plt.bar(x_corr+0.5,[20,10,12],width=0.5,label="Next Year")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,25)
plt.legend()
plt.show()
```

### Pie Chart

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20,10,15,5]
plt.title("Subjects")
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),shadow=True,autopct="%1.1f%%")
plt.show()
```
